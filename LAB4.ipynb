```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def clean_text(text):
    if isinstance(text, str):
        html = re.compile('[<,#*?>]')
        text = html.sub(r'', text)
        text = re.sub(r'[^\u0600-\u06FF\s]+', '', text)
    elif isinstance(text, BeautifulSoup):
        text = ''.join(text.stripped_strings)
        text = re.sub(r'[^\u0600-\u06FF\s]+', '', text)

    return text
```

```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017')
db = client.ArabicNewsletter

collection = db.ArticlesLab4
collection.insert_many(articles)
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer

arabic_stopwords = set(stopwords.words('arabic'))
documents = collection.find()

docAfterTraitment = []

for document in documents:
    #Tokenization
    tokenized_content = word_tokenize(document['content'])
    #Stop words
    filtered_content = [word for word in tokenized_content if word not in arabic_stopwords]
    # Stemming
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(word) for word in filtered_content]
     # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in stemmed_tokens]
    #Discretization
    vectorizer = CountVectorizer(binary=True)
    features = vectorizer.fit_transform(lemmatized_tokens).toarray()
    document['content'] = features
    docAfterTraitment.append(document)
    
```

```python
for document in docAfterTraitment:
    print(document)
```
